```python
import cv2
import numpy as np
```

```python
# Load YOLOv4 model and configuration
net = cv2.dnn.readNet("yolov4.weights", "yolov4.cfg")
```

```python
# Load COCO class labels
classes = []
with open('coco.names') as f:
  class_names = [line.strip() for line in f.readlines()]
```

```python
# Function to perform fruit detection in real-time
def detect_fruits(image):
  # Convert image to blob format
    blob = cv2.dnn.blobFromImage(image, 1.0/255.0, (416, 416), (0,0,0), swapRB=True, crop=False)

    # Set input to the network
    net.setInput(blob)

    # Get output layer indices
    output_layers_indices = net.getUnconnectedOutLayers()
    
    # Get output layer names
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in output_layers_indices]

    # Forward pass through the network
    outputs = net.forward(output_layers)

    # Process detections
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            print(scores)
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:  # Confidence threshold
                # Get bounding box coordinates
                center_x = int(detection[0] * image.shape[1])
                center_y = int(detection[1] * image.shape[0])
                w = int(detection[2] * image.shape[1])
                h = int(detection[3] * image.shape[0])

                # Rectangle coordinates
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                # Draw bounding box and label
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
                #cv2.putText(image, classes[class_id], (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    return None
```

```python
# Main function for real-time detection
def main():
    image = cv2.imread('a2.jpg')
    detected_frame = detect_fruits(image)
    # cap = cv2.VideoCapture(0)  # Use default camera
    # while True:
    #     ret, frame = cap.read()
    #     if not ret:
    #         break

    #     # Perform fruit detection
    #     detected_frame = detect_fruits(frame)
    if detected_frame is not None:
        cv2.imshow('Fruit Detection', detected_frame)
        # if cv2.waitKey(1) & 0xFF == ord('q'):
        #     break
    else:
        print("No objects detected.")

    #     # Display the resulting frame
    #     cv2.imshow('Fruit Detection', detected_frame)
    #     if cv2.waitKey(1) & 0xFF == ord('q'):
    #         break

    # # Release capture and close windows
    # cap.release()
    # cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
